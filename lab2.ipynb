```python
import random as rn
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
N = 100
for i in range(N):
    exp_1 = rn.expovariate(1)
    exp_2 = rn.expovariate(1)
    norm = rn.normalvariate(0,1)
    X.append(np.array([exp_1,exp_2,norm]))
```

```python
plt.hist([X[i][0] for i in range(N)])
plt.title("Експоненційно розподілена з параметром 1 перша координата вектора")
plt.show()
plt.hist([X[i][1] for i in range(N)])
plt.title("Експоненційно розподілена з параметром 1 друга координата вектора")
plt.show()
plt.hist([X[i][2] for i in range(N)])
plt.title("Стандартно нормально розподілена третя координата")
plt.show()
```

```python
alpha = X[0]
while(True): 
    dot_product = []
    for i in range(N):
        dot_product.append(np.dot(alpha,X[i]))

    min = np.min(dot_product)
    argmin = np.argmin(dot_product)

    if min > 0:
        print(alpha, "шуканий розділяючий вектор")
        break
    else:
        k_hat = (np.argmin([np.linalg.norm((1-k)*alpha + k*X[argmin]) for k in np.arange(0.01,1.01,0.01)])+1)/100
        alpha = (1-k_hat)*alpha + k_hat*X[argmin]
        print(alpha, "нешуканий вектор")
```
